# A Simple Movie Recommendation System Logic

### Tasks

- Change genre strings into vectors using TF-IDF Vectorizer
- Compare movies using genre codes through cosine similarity
- Create recommendation function and show it through chart

### Content Filtering Using Genre

```python
df.head()
```

### Here we use titles and their genres for this recommendation logic

```python
# Converting genres column data type into string and checking

df['genres'] = df['genres'].astype(str)

print(df['genres'].dtype)
```

```python
# Combine 'listed_in' (genres) into a single string for each movie

df['genres'] = df['genres'].str.replace(',', '').str.split().str.join(' ')

# Display the processed data

df[['title', 'genres']].head()
```

```python
# Create a TF-IDF Vectorizer to convert genre strings to vectors
# It helps in transforming text into numerical values that can be used and processed by ML algorithms.

tfidf = TfidfVectorizer(stop_words='english')  # The stop words is used to ignore certain words like 'the' or 'and' 

# Fit and transform the data

tfidf_matrix = tfidf.fit_transform(df['genres'])

# Display the shape of the matrix

tfidf_matrix.shape
```

### Finding Similar Movies using Cosine Similarity

- This checks the genre codes for each title and sees how close they are to each other

```python
# Calculate the cosine similarity matrix

cosine_s = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Display the similarity matrix

cosine_s.shape
```

### Recommending Movies and Visualizing them

```python
# Creates a pandas Series where the index is the movie titles (df['title']) and the values are the indices of the DataFrame (df.index).

indices = pd.Series(df.index, index=df['title'])

# Function to recommend movies based on content similarity

def recommend_movies(title, cosine_s=cosine_s):
    
    try:
        
        idx = indices[title]
        sim_scores = cosine_s[idx]
        sorted_indices = sim_scores.argsort()[::-1]
        top_8_indices = sorted_indices[1:9]
        return df['title'].iloc[top_8_indices]
    
    except KeyError:
        
        print(f"Error: The movie '{title}' is not found in the dataset.")
        return []
    
    except Exception as e:
        
        print(f"An unexpected error occurred in recommend_movies: {e}")
        return []

# Test the recommendation system with a sample movie

recommendations = recommend_movies('Ganglands')

recommendations
```

```python
# Visualize content-based recommendations for a given title

def visualize_recom(title, recommendations):
    
    plt.figure(figsize=(10, 6))
    plt.barh(recommendations[::-1], range(1, len(recommendations) + 1), color='powderblue')
    plt.xlabel('Similarity Rank')
    plt.title(f'Movies similar to "{title}"')
    plt.gca().invert_yaxis()
    plt.show()
    
# Visualize recommendations for a sample movie

recommendations = recommend_movies('Ganglands')

visualize_recom('Ganglands', recommendations)
```

### Collaboration Filtering With SVD

- Simulating a user-item interaction matrix and performing SVD to implement collaborative filtering.

```python
# Step 1: Simulate a user-item interaction matrix
# Let's assume we have 100 users and 500 movies

users = 100

movies = df.shape[0]


# Generate a random user-movie rating matrix (ratings between 1 and 5)

np.random.seed(42)  # For reproducibility

user_ratings = np.random.randint(1, 6, size=(users, movies))
```

```python
# Step 2: Perform SVD on the user-item matrix

svd = TruncatedSVD(n_components=20, random_state=42)  # Reduce to 20 latent factors

user_matrix = svd.fit_transform(user_ratings)

movie_matrix = svd.components_.T
```

```python
# Function to recommend movies based on collaborative filtering

def recomm_movies_collab_filter(user_id, user_matrix, movie_matrix, n_recommendations=8):
    
    try:
        
        # Predict ratings by multiplying user and movie matrices
        
        predicted_ratings = np.dot(user_matrix[user_id], movie_matrix.T)
        
        
        # Sort the movies by predicted rating
        
        top_movie_indices = np.argsort(predicted_ratings)[-n_recommendations:][::-1]
        
        
        return df['title'].iloc[top_movie_indices]
    
    
    except IndexError:
        
        print(f"Error: User ID {user_id} is out of bounds.")
        
        return pd.Series([])

    
# Example usage for user with ID 5

collab_recommendations = recomm_movies_collab_filter(5, user_matrix, movie_matrix)

visualize_recom('User 5 Recommendations', collab_recommendations)
```

# Hybrid Recommendation System

### So far, we have constructed:-

- **Content Filtering Recommendation System**
    - Utilizing TFIDF Vectorization 
    - Utilizing cosine similarity
    

- **Collaboration Filtering Recommendation System**
    - Simulated a User-Item interaction matrix
    - Used Truncated Singular Value Decomposition (SVD)

### Now we combine these through Hybrid Recommendations function

```python
# Hybrid Recommendation Function that combines the two methods

def hybrid_recommendations(title, user_id, content_weight=0.5, collaborative_weight=0.5):
    
    try:
        
        # Get content-based recommendations
        
        content_recs = recommend_movies(title)
        
        
        # Get collaborative filtering recommendations
        
        collaborative_recs = recomm_movies_collab_filter(user_id, user_matrix, movie_matrix)
        
        
        # Combine the two sets of recommendations without duplicates
        
        combined_recs = pd.concat([content_recs, collaborative_recs]).drop_duplicates()
        
        
        # Create weighted scores for each recommendation list
        
        content_scores = pd.Series(content_weight, index=content_recs)
        
        collaborative_scores = pd.Series(collaborative_weight, index=collaborative_recs)
        
        
        # Combine the scores, add up the weights
        
        total_scores = content_scores.add(collaborative_scores, fill_value=0)
        
        
        # Sort movies by total score in descending order
        
        total_scores = total_scores.sort_values(ascending=False)
        
        
        # Return the top 8 recommendations
        
        
        return total_scores.index[:8]
    
    except Exception as e:
        
        print(f"An error occurred in hybrid_recommendations: {e}")
        
        return pd.Series([])
```

# User Feedback Mechanism

```python
def get_user_feedback_and_recommend(hybrid_recs, user_id, title, content_weight, collaborative_weight):
    
    updated_recs = hybrid_recs
    
    for movie in hybrid_recs:
        
        while True:  # Loop until valid input is received
            
            user_feedback = input(f"Do you like this recommendation: '{movie}'? (yes/no): ").strip().lower()
            
            if user_feedback == 'no':
                # Reduce content weight and increase collaborative weight
                content_weight -= 0.1
                collaborative_weight += 0.1

                # Generate new recommendations based on updated weights
                updated_recs = hybrid_recommendations(title, user_id, content_weight, collaborative_weight)
                
                print(f"\nUpdated Hybrid Recommendations (Content Weight: {content_weight}, Collaborative Weight: {collaborative_weight}):")
                print(updated_recs)
                
                break  # Exit the loop after getting valid feedback

            elif user_feedback == 'yes':
                print("Great! Enjoy your recommendations.")
                break  # Exit the loop after getting valid feedback

            else:
                print("Invalid input. Please enter 'yes' or 'no'.")
    
    return updated_recs
```

```python
# Example usage for hybrid recommendation system

title = 'Kung Fu Panda'
user_id = 0
content_weight = 0.5
collaborative_weight = 0.5

hybrid_recs = hybrid_recommendations(title, user_id, content_weight, collaborative_weight)

print("Initial Hybrid Recommendations:")

print(hybrid_recs)

# Feedback loop and updated recommendations

hybrid_recs = get_user_feedback_and_recommend(hybrid_recs, user_id, title, content_weight, collaborative_weight)


# Visualizing the final recommendations

visualize_recom(title, hybrid_recs)
```
